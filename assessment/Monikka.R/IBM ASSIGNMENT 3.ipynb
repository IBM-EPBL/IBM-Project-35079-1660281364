```python
data = data[['M', 'F', 'I',	'Length',	'Diameter',	'Height',	'Whole weight',	'Shucked weight',	'Viscera weight',	'Shell weight',	'Rings']]
```

**Scale the independent
variables**

```python
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import StandardScaler
trans = StandardScaler()
x = trans.fit_transform(x)
```

```python
x = pd.DataFrame(x)
```

```python
x.head()
```

**Split the data into training and testing**

```python
from sklearn import model_selection 
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y, test_size=0.33, random_state=42)


print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

**Build the Model, Train the Model, Test the Model, Measure the performance using Metrics.**

```python
print(x_train)
```

```python
from sklearn.metrics import *
from sklearn.tree import DecisionTreeClassifier
def decisionTree(x_train, x_test, y_train, y_test):
    dt=DecisionTreeClassifier() 
    dt.fit(x_train,y_train) 
    yPred = dt.predict(x_test) 
    print(confusion_matrix(y_test,yPred)) 
    print('Classification report') 
    print(classification_report (y_test, yPred))
    print('***DecisionTreeClassifier****')
```

```python
 from sklearn.ensemble import RandomForestClassifier
 from sklearn.metrics import *
def randomForest (x_train, x_test, y_train, y_test):
   rf = RandomForestClassifier() 
   rf.fit(x_train,y_train) 
   yPred = rf.predict(x_test) 
   print ('***RandomForestClassifier****') 
   print('Confusion matrix') 
   print(confusion_matrix(y_test,yPred)) 
   print('Classification report')
   print(classification_report (y_test,yPred))
```

```python
decisionTree(x_train,x_test,y_train,y_test)
```

```python
randomForest (x_train,x_test,y_train,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```
